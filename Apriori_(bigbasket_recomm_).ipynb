```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Big_Basket.csv', header = None)
transactions = []

for i in range (0,200):
  transactions.append([str(dataset.values[i,j]) for j in range(0,9)])
```

```python
print(dataset)
```

```python
print(transactions)
```

```python
from apyori import apriori
basket_intelligence = apriori(transactions = transactions, min_support = 0.001, min_confidence = 0.01, min_lift = 0.5, min_length = 2)
results = list(basket_intelligence)
```

```python
results
```

```python
import pandas as pd

def inspect(results):
    product1 = []
    product2 = []
    supports = []
    confidences = []
    lifts = []

    for result in results:
        # Support for each rule
        support = result.support

        # Extract all ordered_statistics for each result
        for stat in result.ordered_statistics:
            items_base = list(stat.items_base)
            items_add = list(stat.items_add)

            # Ensure both items_base and items_add contain valid items
            if items_base and items_add:
                product1.append(', '.join(items_base))  # Join items if there are multiple
                product2.append(', '.join(items_add))   # Join items if there are multiple
                supports.append(support)
                confidences.append(stat.confidence)
                lifts.append(stat.lift)

    # Create DataFrame with the valid item pairs
    return pd.DataFrame(
        list(zip(product1, product2, supports, confidences, lifts)),
        columns=['product1', 'product2', 'Support', 'Confidence', 'Lift']
    )

# Run the function on the results and display the output
DataFrame_intelligence = inspect(results)
print(DataFrame_intelligence)
```

```python
DataFrame_intelligence.nlargest(n = 10, columns = 'Lift')
```
